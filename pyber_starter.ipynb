```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(city_df, ride_df, on="city")
# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

combined_gp = combined_df.groupby(["type", "city"])

fare_df = combined_gp.mean()
fare_df = fare_df[["fare"]]
fare_df = fare_df.reset_index()


rides_df = combined_gp.count()
rides_df = rides_df[["ride_id"]]
rides_df = rides_df.reset_index()

drivers_df = combined_gp.sum()
drivers_df = drivers_df[["driver_count"]]
drivers_df = drivers_df.reset_index()

merge_df = pd.merge(fare_df, rides_df, how="outer", on=["type","city"])
merge_df = pd.merge(merge_df, drivers_df, how="outer", on=["type", "city"])
merge_df = merge_df.rename(columns={"fare":"Average Fare", "ride_id":"Number of Rides", "driver_count":"Driver Count"})

rural_df = merge_df.loc[merge_df["type"]=="Rural"]
suburban_df = merge_df.loc[merge_df["type"]=="Suburban"]
urban_df = merge_df.loc[merge_df["type"]=="Urban"]

# Build the scatter plots for each city types
ax1 = rural_df.plot(kind="scatter", x="Number of Rides", y="Average Fare", grid=True, figsize=(20,10), title="Pyber Ride Sharing Data (2016)", marker="o", c='y', edgecolors="black", s=merge_df["Driver Count"], label="Rural")
ax2 = suburban_df.plot(ax=ax1, kind="scatter", x="Number of Rides", y="Average Fare", grid=True, figsize=(20,10), title="Pyber Ride Sharing Data (2016)", marker="o", c='b', edgecolors="black", s=merge_df["Driver Count"], label="Suburban")
ax3 = urban_df.plot(ax=ax1, kind="scatter", x="Number of Rides", y="Average Fare", grid=True, figsize=(20,10), title="Pyber Ride Sharing Data (2016)", marker="o", c='r', edgecolors="black", s=merge_df["Driver Count"], label="Urban")

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare_gp = combined_df.groupby("type")
total_fare_df = total_fare_gp.sum()
total_fare_df.head()
# Build Pie Chart

city_type = ["Rural", "Suburban", "Urban"]
total_fare = total_fare_df["fare"]
color = ["yellow","blue","red"]
explode = (0,0,0.1)

plt.pie(total_fare, explode=explode, labels=city_type, colors=color,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride_gp = combined_df.groupby("type")
total_ride_df = total_ride_gp.count()
total_ride_df.head()
# Build Pie Chart

city_type = ["Rural", "Suburban", "Urban"]
total_ride = total_ride_df["city"]
color = ["yellow","blue","red"]
explode = (0,0,0.1)

plt.pie(total_ride, explode=explode, labels=city_type, colors=color,
        autopct="%1.1f%%", shadow=True, startangle=140)


# Save Figure
```
